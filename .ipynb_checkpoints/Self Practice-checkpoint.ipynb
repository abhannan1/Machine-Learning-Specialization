```python
import numpy as np
```

```python
x_train = np.array([1,4,5,6])
y_train = np.array([400,700,900,1000])
```

```python
def LR_model(x, w, b):
    f"""
    Args:
        x (ndarray (m,)) = training data, m examples
        w,b (scaler) = model parameters
    Returns:
        f_wb (ndarray (m,)) = model predictions
    
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

```python
pred = LR_model(x_train, w=200, b=100)
```

```python
pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_train, pred, c="b", label="Predicted Values" )
plt.scatter(x_train, y_train, marker="x", c="r", label = "Actual Values")
plt.title("Housing Prices")
plt.ylabel("Prices (in 1000s of dollars)")
plt.xlabel("Size of the House (1000 sqft)")
plt.legend()
plt.show()
```

```python
np.argmin(pred)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./weight-height.csv')
```

```python
df.head(2)
```

```python
x = df[["Height"]].values
y = df[["Weight"]].values
```

```python
def MSE(y_pred, y):
    mse = (y - y_pred)**2 
    return np.sqrt(mse.mean())
```

```python
def reg_model(x,w,b):
    y = w*x + b 
    return y
```

```python
df.plot(kind="scatter",
        x="Height",
       y= "Weight")
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(10,5))
ax1 = plt.subplot(121)
df.plot(kind="scatter",
       x="Height",
       y="Weight",
       title="Weight and Height in adults",
       ax=ax1)

mses = []
bs = [-150,-100,-50,0,50,100,150]

for b1 in bs:
    y_pred = reg_model(x,w=2,b=b1)
    mse = MSE(y_pred, y)
    mses.append(mse)
    plt.plot(x,y_pred)
    
ax2 = plt.subplot(122)
plt.plot(b,mses, "-o")
plt.xlabel("b")
plt.title("cost as the function of b")
```
